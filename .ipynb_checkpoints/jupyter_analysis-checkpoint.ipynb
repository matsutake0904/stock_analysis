```python
def moving_ave(period, ma_per, num):
    for i in range(period):
        if i ==0:
            ma_df = stock_df[stock_df['num']==num].sort_values('date', ascending=True).iloc[-i-ma_per-1:-i-1].mean()
            date_df = pd.DataFrame([stock_df[stock_df['num']==num].sort_values('date', ascending=True).iloc[-i-1].date], index=['date'])
            return_df = pd.concat([ma_df, date_df], axis=0)
        else:
            ma_df = stock_df[stock_df['num']==num].sort_values('date', ascending=True).iloc[-i-ma_per-1:-i-1].mean()
            date_df = pd.DataFrame([stock_df[stock_df['num']==num].sort_values('date', ascending=True).iloc[-i-1].date], index=['date'])
            temp_df=pd.concat([ma_df, date_df], axis=0)
            return_df = pd.concat([return_df, temp_df], axis=1)
    return return_df.T
```

```python
def period_benefit(num, period,  start_type):
    valid_list = ['open', 'high', 'low', 'close']
    if not start_type in valid_list :
        print('Error arguments 3 and 4 must be in open high low close')
        return False
    
    temp = stock_df[stock_df['num']==num].copy()
    temp['diff']= stock_df[stock_df['num']==num].sort_values('date', ascending=False)[start_type].diff(-period)
    # stock_df[stock_df['num']==4027].sort_values('date', ascending=False)

    # temp.sort_values('date', ascending=False)
    return temp.sort_values('date', ascending=False)
```

```python
def period_benefit_max(num, period,  start_type, end_type):
    valid_list = ['open', 'high', 'low', 'close']
    if not start_type in valid_list :
        print('Error arguments 3 and 4 must be in open high low close')
        return False
    
    temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
    temp['up'] = 0
    temp['up_ratio'] = 0
    temp['down'] = 0
    temp['down_ratio'] = 0

    for i in range(len(temp)-period):
#         print('len df = {}, i={}'.format(len(temp), i))
        max_up = 0
        max_down = 0
        for j in range(period):
            diff = temp[end_type].iloc[i+j] -temp[start_type].iloc[i]
            if max_up < diff:
                max_up = diff
            if max_down > diff:
                max_down = diff
        
        temp['up'].iloc[i] = max_up
        temp['up_ratio'].iloc[i] = max_up / temp[start_type].iloc[i]
        temp['down'].iloc[i] = max_down
        temp['down_ratio'].iloc[i] = max_down / temp[start_type].iloc[i]
#         raise Exception
        
        # temp.sort_values('date', ascending=False)
    return temp.sort_values('date', ascending=False)
```

```python
def get_limit(num, _type):
    valid_list = ['high', 'low', 'yorazu_high', 'yorazu_low']
    if not _type in valid_list :
        print('Error arguments 3 and 4 must be in open high low close')
        return False
    if _type == 'high':
        temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
        temp['diff'] = temp['close'].diff()
        return temp[(temp['high']==temp['close']) & (temp['diff'] > temp['open']*0.1)]
    
    elif _type == 'low':
        temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
        temp['diff'] = temp['close'].diff()
        return temp[(temp['low']==temp['close']) & (temp['diff'] < -temp['open']*0.1)]
    
    elif _type == 'yorazu_high':
        temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
        temp['diff'] = temp['close'].diff()
        return temp[(temp['open']==temp['high'])& (temp['high'] == temp['close']) & (temp['close'] == temp['low'] )& (temp['diff'] > temp['open']*0.1)]
    
    elif _type == 'yorazu_low':
        temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
        temp['diff'] = temp['close'].diff()
        return temp[(temp['open']==temp['high'])& (temp['high'] == temp['close']) & (temp['close'] == temp['low'] )& (temp['diff'] < -temp['open']*0.1)]
                          
```

```python
def get_valiance(num, _type, rate):
    valid_list = ['high', 'low', 'yorazu_high', 'yorazu_low']
    if not _type in valid_list :
        print('Error arguments 3 and 4 must be in open high low close')
        return False
    if _type == 'high':
        temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
        temp['diff'] = temp['close'].diff()
        return temp[(temp['diff'] > temp['open']*rate)]
    
    elif _type == 'low':
        temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
        temp['diff'] = temp['close'].diff()
        return temp[ (temp['diff'] < -temp['open']*rate)]
    
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def plot_stock_data(num):
    temp =stock_df[stock_df['num']==num].copy().sort_values('date', ascending=True)
    cmap = ListedColormap(['#0000FF', '#FF0000', '#00FF00'])
    list_tmp = ['open', 'high', 'low']
    counter = 0
    for i in list_tmp:
        counter += 1
        if i == 'open':
            temp['col'] = counter -1
            x = temp['date']
            y = temp['open']
            c = temp['col']
        else:
            limit = get_valiance(num, i, 0.1)
            limit['col'] = counter - 1
            x = x.append(limit['date'])
            y = y.append(limit['close'])
            c = c.append(limit['col'])
            print( limit['date'], limit['close'], limit['col'])
            
    ax = plt.subplot(1, 1, 1)
    ax.scatter(x, y, c=c, cmap=cmap)
    plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np
```

```python
def plot_chart(num):

    df = stock_df[stock_df['num']==num]
    df = df.sort_values('date', ascending=True).set_index('date')
    x = np.arange(df.shape[0])    
    
    df2 = get_valiance(num, 'low' ,0.1).sort_values('date', ascending=True)
    x2=[]
    for i in range(len(df2)):
        try:
            x2.append(df.index.get_loc(df2['date'].iloc[i]))
        except Exception as e:
            print("error {}".format(e))
    df2.set_index('date')
    data2 =go.Scatter(
                    x = x2,
                    y=df2['close'],
                    mode='markers',
                    marker=dict(
                    size=8,
                    color='rgb({}, {}, {})'.format(125, 255, 255),
#                     colorscale='deep',
                    showscale=False),
                    name='low')

    df2 = get_valiance(num, 'high' ,0.1).sort_values('date', ascending=True)
    x2=[]
    for i in range(len(df2)):
        try:
            x2.append(df.index.get_loc(df2['date'].iloc[i]))
        except Exception as e:
            print("error {}".format(e))
    df2.set_index('date')
    data3 =go.Scatter(
                    x = x2,
                    y=df2['close'],
                    mode='markers',
                    marker=dict(
                    size=8,
                    color='rgb({}, {}, {})'.format(255, 125, 125),
#                     colorscale='deep',
                    showscale=False),
                    name='high')

    
    
    interval = 100
    vals = np.arange(df.shape[0], step=interval)
#     vals = [(df['date'].iloc[0] + num/interval*(df['date'].iloc[-1] - df['date'].iloc[0])).strftime("%Y-%m-%d") for num in range(10) ]
    print(vals)
    labels = list(df[::interval].index)
    
    fig = go.Figure(
            data=[go.Candlestick(
                    x=x,
                    open=df['open'],
                    high=df['high'],
                    low=df['low'],
                    close=df['close'],
                    name=str(num)),
                  data2,
                  data3

                 ],
            layout = go.Layout(
                    xaxis = dict(
                        tickvals = vals,
                        ticktext = labels,
                        tickangle=-45 
                    ),
            )
    )
    fig.show()
```
